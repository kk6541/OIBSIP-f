# Pre- Processing  the Data

-> By using the label encoder ,We can convert object values to numerical values

```python
from sklearn.preprocessing import LabelEncoder
object_data = dataset.select_dtypes(include='object')
num_data = dataset.select_dtypes(exclude='object')

enc = LabelEncoder()
for i in range(0, object_data.shape[1]):
    object_data.iloc[:,i] = enc.fit_transform(object_data.iloc[:,i]) 
```

```python
object_data.info()
```

```python
num_data.info()
```

```python
data = pd.concat([num_data, object_data], axis = 1)
```

```python
data.info()
```

```python
data
```

# Train-test-spilt the data 


```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(columns="price")           
y = data["price"]  
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7)
```

```python
print("x Train : ", x_train.shape)
print("x Test  : ", x_test.shape)
print("Y Train : ", y_train.shape)
print("Y Test  : ", y_test.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(x_train,y_train)
predictions_test=model.predict(x_test)
print(model.score(x_test,y_test))
```
